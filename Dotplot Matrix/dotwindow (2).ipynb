```python

# create matrix given dimensions: number of rows and columns filled with zeros
def create_mat(nrows, ncols):
    mat = []
    for i in range(nrows):
        mat.append([])
        for j in range(ncols):
            mat[i].append(0)
    return mat

# basic dotplot algorithm: fills with ones coincident characters
def dotplot(seq1, seq2):
    mat = create_mat(len(seq1), len(seq2))
    for i in range(len(seq1)):
        for j in range(len(seq2)):
            if seq1[i] == seq2[j]:
                mat[i][j] = 1
    return mat

# extended dotplot with window and threshold parameters
def extended_dotplot (seq1, seq2, window,step, threshold):
    mat = create_mat(len(seq1), len(seq2))
    start = int(window/2)
    for i in range(start,len(seq1)-start,step):
        for j in range(start, len(seq2)-start,step):
            matches = 0
            l = j - start 
            for k in range(i-start, i+start+1):
                if seq1[k] == seq2[l]: matches += 1
                l += 1
                if matches >= threshold: mat[i][j] = 1

    return mat

# prints dotplot
def print_dotplot(mat, s1, s2):
    import sys
    sys.stdout.write(" " + s2+"\n")
    for i in range(len(mat)):
        sys.stdout.write(s1[i])
        for j in range(len(mat[i])):
            if mat[i][j] >= 1:
                sys.stdout.write("*")
            else:
                sys.stdout.write(" ")
        sys.stdout.write("\n")


def test():
    s1 = "ACCTTGTCCTCTTTGCCC"
    s2 = "ACGTTGACCTGTAACCTC"
    mat1 = dotplot(s1, s2)
    print_dotplot(mat1, s1, s2)
    
    print("")    
    mat2 = extended_dotplot(s1, s2, 9,3, 4)
    print_dotplot(mat2, s1, s2)
    

if __name__ == "__main__": 
    test()
```
